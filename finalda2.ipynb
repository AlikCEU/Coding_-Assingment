```python
import pandas as pd
data = pd.read_csv('C:/Users/PC/Desktop/alik/DA2/games_may2024.csv')
```

```python
main_analysis = ['estimated_owners', 'price', 'metacritic_score', 'genres', 'pct_pos_total', 'dlc_count','required_age']
```

```python
data_main_analysis = data[main_analysis]
data_main_analysis
```

```python
data_main_analysis .dtypes
```

```python
data_main_analysis[['owners_min', 'owners_max']] = data_main_analysis['estimated_owners'].str.split(' - ', expand=True)
data_main_analysis['owners_min'] = pd.to_numeric(data_main_analysis['owners_min'], errors='coerce')
data_main_analysis['owners_max'] = pd.to_numeric(data_main_analysis['owners_max'], errors='coerce')
data_main_analysis['estimated_owners'] = (data_main_analysis['owners_min'] + data_main_analysis['owners_max']) / 2
data_main_analysis['estimated_owners'] = data_main_analysis['estimated_owners'].astype('float64')
data_main_analysis.drop(['owners_min', 'owners_max'], axis=1, inplace=True)
print(data_main_analysis[['estimated_owners']].head())
```

```python
import ast
data_main_analysis['genres'] = data_main_analysis['genres'].apply(ast.literal_eval)
data_main_analysis = data_main_analysis.dropna()
all_genres = data_main_analysis['genres'].explode().unique()
for genre in all_genres:
    data_main_analysis[genre] = data_main_analysis['genres'].apply(lambda x: 1 if genre in x else 0)
print(data_main_analysis.head())
```

```python
data_main_analysis
```

```python
missing_values = data_main_analysis.isnull().sum()
data_main_analysis.dropna(inplace=True)
print(missing_values)
```

```python
variables = ['estimated_owners','price', 'metacritic_score', 'pct_pos_total', 'dlc_count']
description = data_main_analysis[variables].describe().T
description['range'] = description['max'] - description['min']
description['skew'] = data_main_analysis[variables].skew()
description['kurtosis'] = data_main_analysis[variables].kurtosis()
pd.options.display.float_format = '{:.3f}'.format
print(description[['mean', 'std', 'min', '25%', '50%', '75%', 'max']])
```

```python
# Replace -1.0 with 0.0 in the pct_pos_total column
data_main_analysis['pct_pos_total'] = data_main_analysis['pct_pos_total'].replace(-1.0, 0.0)

print(data_main_analysis['pct_pos_total'].describe())
```

```python
data_filtered  = data_main_analysis
data_filtered = data_filtered[data_filtered['estimated_owners'] > 0]
print(data_filtered.shape)
```

```python
import seaborn as sns
import numpy as np
from scipy import stats

data_filtered['boxcox_metacritic_score'], lambda_metacritic = stats.boxcox(data_filtered['metacritic_score'] + 1)
y = np.log(data_filtered['estimated_owners'] + 1) 
X = data_filtered[['price', 'boxcox_metacritic_score']]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())

price_grid = np.linspace(data_filtered['price'].min(), data_filtered['price'].max(), 100)
mean_boxcox_metacritic = data_filtered['boxcox_metacritic_score'].mean()

predicted_log_owners = model.params[0] + model.params[1] * price_grid + model.params[2] * mean_boxcox_metacritic

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_filtered['price'], y=y, color='blue', alpha=0.5, label="Actual Data")

plt.plot(price_grid, predicted_log_owners, color='red', label="Fitted Regression Line")
plt.title('Multiple Regression: Price and Box-Cox Metacritic Score vs Log-Transformed Estimated Owners', fontsize=14)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Log-Transformed Estimated Owners', fontsize=12)
plt.legend()

plt.grid(True)
plt.show()
```

```python
data_group_1 = data_filtered[data_filtered['required_age'] <= 15]
data_group_2 = data_filtered[data_filtered['required_age'] > 15]

independent_vars = ['price', 'boxcox_metacritic_score', 'Action', 'Free to Play', 
                   'Adventure', 'Massively Multiplayer', 'Strategy', 'Accounting', 
                   'sqrt_dlc_count', 'pct_pos_total']

X_group_1 = data_group_1[independent_vars]
X_group_1 = sm.add_constant(X_group_1)
Y_group_1 = data_group_1['log_estimated_owners']

model_group_1 = sm.OLS(Y_group_1, X_group_1).fit()

X_group_2 = data_group_2[independent_vars]
X_group_2 = sm.add_constant(X_group_2) 
Y_group_2 = data_group_2['log_estimated_owners']

model_group_2 = sm.OLS(Y_group_2, X_group_2).fit()

print("Regression Results for Age Group 0 to 15:")
print(model_group_1.summary())

print("\nRegression Results for Age Group 15 and above:")
print(model_group_2.summary())
```

```python
from sklearn.model_selection import train_test_split
X = data_filtered[['price', 'boxcox_metacritic_score', 'Action', 'Free to Play', 
                   'Adventure', 'Massively Multiplayer', 'Strategy', 'Accounting', 
                   'sqrt_dlc_count', 'pct_pos_total']]

X = sm.add_constant(X)
Y = data_filtered['log_estimated_owners']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model_train = sm.OLS(Y_train, X_train).fit()
print("Training Set Results:")
print(model_train.summary())
Y_pred = model_train.predict(X_test)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print("\nTest Set Performance:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared: {r2}")
```
